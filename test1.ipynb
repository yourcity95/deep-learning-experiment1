```python
#取数据集
from sklearn.datasets import load_svmlight_file

def get_data():
    data = load_svmlight_file("Housing_scale")
    return data[0], data[1]

X, y = get_data()
```

```python
#分割数据集
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 )
```

```python
#初始化数据
W=np.zeros(13)
b=0
```

```python
#迭代更新
import matplotlib.pyplot as plt
it=100
plt.figure(1)
L=np.zeros(it)
L0=np.zeros(it)
for i in range(it):
    y_pre=X_train*W.T+b
    #计算loss值
    L[i]=1/2*sum((y_pre-y_train)**2)
    #计算梯度的相反数
    D=-(y_pre-y_train).T*X_train
    #更新权值
    W=W+0.001*D
    b=b-0.001*np.sum(y_pre-y_train)
    #预测验证集
    y_pre_test=X_test*W.T+b
    #计算验证集loss
    L0[i]=1/2*sum((y_pre_test-y_test)**2)
#输出最终loss
print(L[-1])
print(L0[-1])
#画出loss的对数随迭代次数的变化
plt.plot(np.log(L),color='r',label='Train')
plt.plot(np.log(L0),label='Test')
plt.title('Linear Regression')
plt.legend()
plt.ylabel('log(loss)')
plt.xlabel('iteration')
plt.show()
```
